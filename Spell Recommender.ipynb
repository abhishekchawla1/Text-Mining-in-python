```python
def jaccard(a,b):
    s1=set(a)
    s2=set(b)
    inter=len(s1.intersection(s2))
    union=len(s1.union(s2))
    return inter/union if union!=0 else 0
```

```python
def spel_rec(input,correct):
    input_bigrams=[input[i:i+2] for i in range(len(input)-1)]
    rec={}
    for word in correct:
        if word[0].lower()==input[0].lower() and len(word)==len(input):
            word_bigram=[word[i:i+2] for i in range(len(word)-1)]
            similarity=jaccard(input,word)
            rec[word]=similarity
        
    sorted_words=sorted(rec.items(),key=lambda x :x[1],reverse=True)
    return sorted_words
    
```

```python
def rec(input,correct):
    l=[]
    for x in input:
        d={}
        s=set(ngrams(x,3))
        for c in correct:
            s2=set(ngrams(c,3))
            if x[0].lower()==c[0].lower():
                d[c]=jaccard_distance(s,s2)
        min_key=min(d,key=d.get)
        l.append(min_key)
    return l
```

```python
#using 4 gram approach

def recc(input,correct):
    l=[]
    for i in input:
        d={}
        s=set(ngrams(i,4))
        for w in correct:
            if w[0].lower()==i[0].lower():
                s2=set(ngrams(w,4))
                d[w]=jaccard_distance(s,s2)
        min_key=min(d,key=d.get)
        l.append(min_key)
    return l
```

```python
recc(input,words)
```

```python
from nltk.metrics import edit_distance
```

```python
edit_distance('rain','shine',substitution_cost=1,transpositions=False)
```

```python
def edit(input,correct):
    l=[]
    for i in input:
        d={}
        for w in correct:
            if w[0].lower()==i[0].lower():
                d[w]=edit_distance(i,w,substitution_cost=1,transpositions=False)
        min_key=min(d,key=d.get)
        l.append(min_key)
    return l
            
```

```python
edit(input,words)
```
