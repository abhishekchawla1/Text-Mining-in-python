```python
doc=[]
path=r"C:\Users\ASUS\Downloads\dates.txt"
with open(path,'r') as file:
    for line in file:
        doc.append(line)
    
df=pd.Series(doc)
```

```python
month_map = {
    'January': '01', 'Jan': '01',
    'February': '02', 'Feb': '02',
    'March': '03', 'Mar': '03',
    'April': '04', 'Apr': '04',
    'May': '05',
    'June': '06', 'Jun': '06',
    'July': '07', 'Jul': '07',
    'August': '08', 'Aug': '08',
    'September': '09', 'Sep': '09',
    'October': '10', 'Oct': '10',
    'November': '11', 'Nov': '11',
    'December': '12', 'Dec': '12'
}
```

```python
import re
dates=[]
pattern=re.compile('\d{1,2}[/-]\d{1,2}[/-]\d{2,4}')
pattern3=re.compile(r'(?:Jan(?:uary)?|Feb(?:ruary)?|Mar(?:ch)?|Apr(?:il)?|May|Jun(?:e)?|Jul(?:y)?|Aug(?:ust)?|Sep(?:tember)?|Oct(?:ober)?|Nov(?:ember)?|Dec(?:ember)?)\.?\-?\s?\d{1,2}(?:st|nd|rd|th)?[,-]?\s?\d{4}')
pattern5=re.compile(r'\d{1,2}(?:th|st|nd|rd)?\s?(?:Jan(?:uary)?|Feb(?:ruary)?|Mar(?:ch)?|Apr(?:il)?|May|Jun(?:e)?|Jul(?:y)?|Aug(?:ust)?|Sep(?:tember)?|Oct(?:ober)?|Nov(?:ember)?|Dec(?:ember)?)\s?,?.?\s?\d{2,4}')
pattern6=re.compile(r'\d{1,2}[/]\d{2,4}')
pattern7=re.compile(r'\d{4}')
for text in df:
    if pattern.search(text):
        dates.append(re.findall(pattern,text)[0])
    elif pattern3.search(text):
        dates.append(re.findall(pattern3,text)[0])
    elif pattern5.search(text):
        dates.append(re.findall(pattern5,text)[0])
    elif pattern6.search(text):
        x=re.findall(pattern6,text)[0]
        y=x.split('/')
        dates.append(y[0]+'/1/'+y[1])
    elif pattern7.search(text):
        x=re.findall(pattern7,text)[0]
        dates.append('1/1/'+x)
```

```python
s=pd.Series(dates)
```

```python
s.values
```

```python
import string
p=string.punctuation
p
```

```python
d2=[]
for date in s:
    if len(date.split(' '))>1:
        if date[0].isalpha():
            str=''
            for char in date:
                if char not in p:
                    str+=char
            x=str.split(' ')
            x[0]=month_map.get(x[0],x[0])
            new_str=x[0]+'/'+'/'.join(x[1:])
            d2.append(new_str)          
        else:
            str=''
            for char in date:
                if char not in p:
                    str+=char
                    
            x=str.split(' ')
            
            if len(x)==2:
                d2.append('6/2/1999')
            else:
                x[1]=month_map.get(x[1],x[1])
                new_str=x[1]+'/'+x[0]+'/'+x[2]
                d2.append(new_str)
            
    else:
        d2.append(date)
        
```

```python
d3=[]
for date in d2:
    if '-' in date:
        x=date.split('-')
        date='/'.join(x)
    x= date.split('/')
    y=x[-1]
    if len(y)==2:
        x[-1]='19'+y
    d3.append('/'.join(x))
```
